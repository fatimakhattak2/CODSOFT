```python
movism=movism[['id', 'title', 'overview', 'genre']]
movism.head()
```

```python
movism['tags'] = movism['title']+movism['genre']
movism.head()
```

```python
final_result  = movism.drop(columns=['overview'])
final_result.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Create a CountVectorizer object with custom parameters
cv = CountVectorizer(
    max_features=10000,  # Maximum number of features
    stop_words='english',  # Stopwords to be removed
    # You can add more parameters here as needed
)

# Fit and transform the 'tags' data into a matrix of token counts
vector = cv.fit_transform(final_result['tags'].values.astype('U')).toarray()

# Check the shape of the resulting matrix
vector_shape = vector.shape
vector_shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vector)
similarity
```

```python
final_result[final_result['title'].str.strip()=="Dilwale Dulhania Le Jayenge"].index[0]
```

```python
distance = sorted(list(enumerate(similarity[1])), reverse=True, key=lambda vector:vector[1])
for i in distance[0:5]:
    print(final_result.iloc[i[0]].title)
```
